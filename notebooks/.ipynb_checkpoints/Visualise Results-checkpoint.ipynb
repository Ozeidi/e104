```python
import json
import itertools
from sklearn import preprocessing
from sklearn.externals import joblib
import numpy as np
import pandas as pd
from datetime import datetime as dt
#Visualisation
#import matplotlib
import matplotlib.pyplot as plt
#%matplotlib inline
%matplotlib notebook
# from pylab import rcParams
# from matplotlib import gridspec
# import seaborn as sns
# from ggplot import *
# import plotly.plotly as py
# #Formatted Print 
# from IPython.display import Markdown, display

#NILMTK
from nilmtk import DataSet
from nilmtk.utils import print_dict
from nilmtk.datastore import HDFDataStore
from nilmtk.electric import align_two_meters



allowed_key_names = ['fridge','microwave','dish_washer','kettle','washing_machine']
WINDOW_PER_BUILDING = {
    1: ("2013-04-12", "2014-12-15"),
    2: ("2013-05-22", "2013-10-03 06:16:00"),
    3: ("2013-02-27", "2013-03-27 06:15:05"),
    4: ("2013-03-09", "2013-04-24 06:15:14"),
    5: ("2014-06-29", "2014-09-01")
}
```

```python
def vis (key_name,model_name, end_e):

    
    sp =60
    # =======  Open configuration file  ===============
    appliances = json.load(open("../conf/apps.json",'r'))
    
    nilmtk_key= appliances[key_name]['key']
    save_path ='../output/{}/{}'.format(model_name,key_name.replace('_',' '))
    v = appliances[key_name]
    data_path='/projects/da33/ozeidi/Project/data/UKDALE/ukdale.h5'
    #training_set = DataSet(data_path)
    validation_set = DataSet(data_path)

    disag_filename =  '{}/{}_disag-out_{}epochs.h5'.format(save_path,nilmtk_key,end_e) # The filename of the resulting datastore
    output = DataSet(disag_filename)

    val_builidng = v['validation_buildings'][0]


    validation_set.set_window(start=WINDOW_PER_BUILDING[val_builidng][0],end=WINDOW_PER_BUILDING[val_builidng][1])
    # This makes a list of references to the NILMTK main meters and submeter
    # no data is generated from them at this point
    mains = validation_set.buildings[val_builidng].elec.mains().power_series_all_data(sample_period=sp)
    meter= dict()
    meter['Pred'] = output.buildings[val_builidng].elec.submeters()[nilmtk_key].power_series_all_data(sample_period=sp)
    for key in allowed_key_names:
        k= appliances[key]['key']
        meter[k] = validation_set.buildings[val_builidng].elec.submeters()[k].power_series_all_data(sample_period=sp)

    # ------Plotting------------
    ax=[None]*6
    fig, ax = plt.subplots(6,1,sharex=True)
    fig.set_size_inches(7, 14)
    plt.xticks(rotation = 90)
    #main plot
    ax[0].plot(mains, label='Mains',color='red')
    ax[0].set_title('Mains Consumption')
    ax[0].set_ylabel('Power (W)')

    # submeter plot
    ax[1].plot( meter[nilmtk_key], label='Ground Truth', color ='green')
    ax[1].plot( meter['Pred'], label='Predicted', color = 'blue')
    ax[1].set_title('{} Predicted and GT'.format(key_name))
    ax[1].set_ylabel('Power (W)')
    for i,key in enumerate(set(allowed_key_names)-set([key_name])):
        i +=2
        k= appliances[key]['key']
        ax[i].plot(meter[k], label ='')
        ax[i].set_title(k)
        ax[i].set_ylabel('Power (W)')
    
    fig.legend(loc='best',fancybox=True, shadow=True, ncol=3)
    fig.suptitle("Comparison Plot",fontsize=14, fontweight='bold')
```

```python
def vis_app (key_name, end_e):
    # =======  Open configuration file  ===============
    appliances = json.load(open("../conf/apps.json"))
    
    nilmtk_key= appliances[key_name]['key']
    v = appliances[key_name]
   
    
    
    save_path ='../output/{}'.format(key_name)
    disag_filename =  '{}/{}_disag-out_{}epochs.h5'.format(save_path,nilmtk_key,end_e) # The filename of the resulting datastore
    data_path='/projects/da33/ozeidi/Project/data/UKDALE/ukdale.h5'

    DS = DataSet(data_path)
    output = DataSet(disag_filename)
    buildings = v['train_buildings'] + v['validation_buildings']
    val_building = v['validation_buildings'][0]

    #mains = DS.buildings[val_builidng].elec.mains().power_series_all_data(sample_period=6)
    meter= dict()
    for i, b in enumerate(buildings):
        DS.set_window(start=WINDOW_PER_BUILDING[b][0],end=WINDOW_PER_BUILDING[b][1])
        meter[b] = DS.buildings[b].elec.submeters()[nilmtk_key].power_series_all_data(sample_period=6)
        

    # This makes a list of references to the NILMTK main meters and submeter
    # no data is generated from them at this point
    meter['Pred'] = output.buildings[val_building].elec.submeters()[nilmtk_key].power_series_all_data(sample_period=6)

    # ------Plotting------------
    ax=[None]*len(buildings)
    fig, ax = plt.subplots(len(ax),1,sharex=True)
    fig.set_size_inches(7, 14)
    plt.xticks(rotation = 90)
    # submeter plot
    ax[0].plot( meter[val_building], label='GT', color ='green')
    ax[0].plot( meter['Pred'], label='Predicted', color = 'blue')
    ax[0].set_title('{} Predicted and GT'.format(key_name))
    ax[0].set_ylabel('Power (W)')
    
    for i,k in enumerate(buildings[:-1]):
        i +=1
        ax[i].plot(meter[k], label ='')
        ax[i].set_title(k)
        ax[i].set_ylabel('Power (W)')
 
    fig.legend(loc='best',fancybox=True, shadow=True, ncol=3)
    fig.suptitle("Comparison Plot",fontsize=14, fontweight='bold')
```

```python
vis_app('kettle',5)
```

```python

def rearrange(cluster_centers, labels):
    centroids = cluster_centers.reshape(1,-1)[0]
    ix = np.argsort(centroids)
    #print(ix)
    lut= np.zeros_like(ix)
    lut[ix]=np.arange(1,len(centroids)+1)
    return (lut[labels])
#rearrang(clusterer.cluster_centers_,clusterer.labels_)

def cluster_series(clusterer, meter_series, on_threshold):
    onehot_encoder = OneHotEncoder(sparse=False)
    meter_series = meter_series.reshape(-1,1)
    ix = meter_series >=on_threshold
    cluster_labels = clusterer.predict(meter_series[ix].reshape(-1,1))
    
    cluster_labels = rearrange(clusterer.cluster_centers_, cluster_labels )
    print(set(cluster_labels))
    meter_series[ix]= cluster_labels

    meter_series[ix==False] = 0
    meter_series =  onehot_encoder.fit_transform(meter_series)
    return meter_series

def lut(clusterer):
    c= np.sort(clusterer.cluster_centers_)
    c = np.append(np.zeros(1),c)
    return c
# if __name__ == '__main__':
#     app='dish_washer'
#     clf =joblib.load('ShortSeq2Point/appconf/{}_clf.pkl'.format(app))
#     y_train = np.load("ShortSeq2Point/dataset/trainsets/Y-{}.npy".format(app))
#     res = cluster_series(clf, y_train,10)
#     print(res)
```
